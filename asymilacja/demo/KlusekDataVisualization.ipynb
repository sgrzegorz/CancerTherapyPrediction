```python
import pandas as pd
import matplotlib.pyplot as plt
from definitions import ROOT_DIR
from asymilacja.utils.datasets import convertIterationsToDays
%cd $ROOT_DIR

def plot(path,plotMtd=False,extrema=None,time=False, title =True):
    header = ['ITERATION', 'ALIVE_CELLS',  'SILENT_CELLS' , 'DEAD_CELLS',  'NUMBER_OF_ARTERY_GIVING_OXYGEN', 'MTD','CUREMENT']
    df = pd.read_csv (path,sep=' ',names=header,skiprows=1)
    x =df['ITERATION']
    xLabel = "nr iteracji"
    if time:
        x = convertIterationsToDays(df['ITERATION'])
        xLabel = 'time [days]'

    plt.plot(x, df['ALIVE_CELLS'] + df['SILENT_CELLS'], label = "komórki nowotworowe")

    # plt.plot(x, df['DEAD_CELLS'], label = "rdzeń nekrotyczny")
    plt.title('Dynamika nowotworu')
    if title:
       plt.title(f'Dynamika nowotworu {path}')


    plt.xlabel(xLabel)
    plt.ylabel('ilość komórek nowotworowych')

    plt.legend()
    plt.show()

    if plotMtd:
        plt.plot(x, df['MTD'], label = "mean tumor volume")
        plt.title('Objętość nowotworu')
        if title:
            plt.title(f'Objętość nowotworu {path}')
        plt.xlabel(xLabel)
        plt.ylabel('volume [mm]')
        plt.show()

# def plot_processed():
#     from data.klusek.patient4.config import all_extrema
#     print(all_extrema)
#
#
#     if extrema is not None:
#         df.loc[df['ITERATION'] in extrema, 'A']
#         plt.plot(extrema, list(df[df.ITERATION == extrema]))
# plot('/home/x/doc/dev/master/results/6/stats0.txt')
```
